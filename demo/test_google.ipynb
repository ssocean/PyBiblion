```python
import requests
from bs4 import BeautifulSoup

import requests
import httpx
# This is the code used to fetch the annual publication count in a specific field.
def get_pubnum(keywords):
    init_kwargs = {"follow_redirects": True}
    _HEADERS = {
        'accept-language': 'en-US,en',
        'accept': 'text/html,application/xhtml+xml,application/xml',
        'User-Agent': 'Mozilla/6.0 (X11; Linux x86_64) AppleWebKit/537.36  Chrome/90.0.3987.149 Safari/537.36',
    }
    url_base = f'https://scholar.google.com/scholar?as_q=%22{keywords}%22'
    # url_base = f'https://scholar.google.com/scholar?as_q=sar+detection'
    years = [i for i in range(2010,2024)]
    # print(years)
    for year in years:
        url = url_base+f'&as_sdt=0%2C5&as_rr=1&as_ylo={year}&as_yhi={year}'
        # self._session.headers.update(_HEADERS)
        # url = 'https://scholar.google.com/scholar?as_q=%22object%20detection%22&as_sdt=0%2C5&as_rr=1&as_ylo=2010&as_yhi=2010'
        print(url)
        init_kwargs.update(headers=_HEADERS)
        session = httpx.Client(**init_kwargs)
        page = session.get(url,timeout=3)
        
        
        soup = BeautifulSoup(page.content, 'html.parser')
        # print(soup)
        # print(soup)
        results = soup.find_all('div', {'class': 'gs_ab_mdw'})
        print(results)
        print(year,results[1].text)
    

keywords = 'object%20detection'
get_pubnum(keywords)
```
